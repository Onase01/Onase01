```python
# Importing Necessary Libraries
from skimage import data
from skimage.color import rgb2gray
import matplotlib.pyplot as plt
import skimage.filters
import numpy as np
 
# Setting the plot size to 15,15
plt.figure(figsize=(15, 15))
 
# Sample Image of scikit-image package
BoilTest = plt.imread("/content/Boil Sample1.jpg")
plt.subplot(1, 2, 1)

 
# Displaying the sample image
plt.imshow(BoilTest)
 
# Converting RGB image to Monochrome
gray_BoilTest = rgb2gray(BoilTest)
plt.subplot(1, 2, 2)
 
# Displaying the sample image - Monochrome
# Format
plt.imshow(gray_BoilTest, cmap="gray")
```

```python
#Calculate the area and number of white and black pixels
# import necessary library
import matplotlib.pyplot as plt
 
# read an image
img = plt.imread("/content/Boil Sample1.jpg")
# displaying the image
plt.imshow(img)
 
# fetch the height and width
height, width, _ = img.shape
 
# area is calculated as height x width
area = height * width
 
# display the area
print("Area of the image is : ", area)



 
```

```python
# blur the image to denoise
blurred_BoilTest = skimage.filters.gaussian(gray_BoilTest, sigma=1.0)
# create a mask based on the threshold
t = 0.4

binary_mask = blurred_BoilTest > t

fig, ax = plt.subplots()
plt.imshow(binary_mask, cmap='gray')
plt.show()
```

```python
number_of_white_pix = np.sum(binary_mask == 255)
number_of_black_pix = np.sum(binary_mask == 0)
  
print('Number of white pixels:', number_of_white_pix)
print('Number of black pixels:', number_of_black_pix)
```

```python
#create a histogram of the blurred grayscale image
histogram, bin_edges = np.histogram(blurred_BoilTest, bins=256, range=(0.0, 1.0))

fig, ax = plt.subplots()
plt.plot(bin_edges[0:-1], histogram)
plt.title("Grayscale Histogram")
plt.xlabel("grayscale value")
plt.ylabel("pixels")
plt.xlim(0, 1.0)
plt.show()
```
