```python
# Importing Necessary Libraries
from skimage import data
from skimage.color import rgb2gray
import matplotlib.pyplot as plt
 
# Setting the plot size to 15,15
plt.figure(figsize=(15, 15))
 
# Sample Image of scikit-image package
BoilTest = plt.imread("/content/Boil Sample.jpg")
plt.subplot(1, 2, 1)
 
# Displaying the sample image
plt.imshow(BoilTest)
 
# Converting RGB image to Monochrome
gray_BoilTest = rgb2gray(BoilTest)
plt.subplot(1, 2, 2)
 
# Displaying the sample image - Monochrome
# Format
plt.imshow(gray_BoilTest, cmap="gray")
```

```python
import skimage.filters
import numpy as np
# blur the image to denoise
blurred_BoilTest = skimage.filters.gaussian(gray_BoilTest, sigma=1.0)
 #create a histogram of the blurred grayscale image
histogram, bin_edges = np.histogram(blurred_BoilTest, bins=256, range=(0.0, 1.0))

fig, ax = plt.subplots()
plt.plot(bin_edges[0:-1], histogram)
plt.title("Grayscale Histogram")
plt.xlabel("grayscale value")
plt.ylabel("pixels")
plt.xlim(0, 1.0)
plt.show()
```

```python
# Importing necessary libraries
from skimage import data
from skimage import filters
from skimage.color import rgb2gray
import matplotlib.pyplot as plt
 
# Setting plot size to 15, 15
plt.figure(figsize=(15, 15))
 
 # Sample Image of scikit-image package
BoilTest = plt.imread("/content/Boil Sample.jpg")
# Converting RGB image to Monochrome
gray_BoilTest = rgb2gray(BoilTest)
 
# Computing Otsu's thresholding value
threshold = filters.threshold_otsu(gray_BoilTest)
 
# Computing binarized values using the obtained
# threshold
binarized_BoilTest = (gray_BoilTest > threshold)*1
plt.subplot(2,2,1)
plt.title("Threshold: >"+str(threshold))
 
# Displaying the binarized image
plt.imshow(binarized_BoilTest, cmap = "gray")
 
# Computing Ni black's local pixel
# threshold values for every pixel
threshold = filters.threshold_niblack(gray_BoilTest)
 
# Computing binarized values using the obtained
# threshold
binarized_coffee = (gray_BoilTest > threshold)*1
plt.subplot(2,2,2)
plt.title("Niblack Thresholding")
 
# Displaying the binarized image
plt.imshow(binarized_BoilTest, cmap = "gray")
 # Computing Sauvola's local pixel threshold
# values for every pixel - Not Binarized
threshold = filters.threshold_sauvola(gray_BoilTest)
plt.subplot(2,2,3)
plt.title("Sauvola Thresholding")
 
# Displaying the local threshold values
plt.imshow(threshold, cmap = "gray")
 
# Computing Sauvola's local pixel
# threshold values for every pixel - Binarized
binarized_coffee = (gray_BoilTest > threshold)*1
plt.subplot(2,2,4)
plt.title("Sauvola Thresholding - Converting to 0's and 1's")
 
# Displaying the binarized image
plt.imshow(binarized_BoilTest, cmap = "gray")
```
